```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import os
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
import seaborn as sns
from sklearn.metrics import f1_score
from sklearn import svm, datasets
from sklearn.metrics import plot_confusion_matrix
from sklearn import metrics

```

```python
data = pd.read_csv("C:\\Users\\DELL\\Desktop\\SVM\\SVMdataset.csv")
data.head()
x= np.asarray(data[['x1', 'x2']])
y=np.asarray(data['y'])
```

```python
# split the dataset into 3 parts

X_train, Zz, y_train, yy = train_test_split(x, y, test_size = 0.4, random_state = 0)

X_validation, X_test, y_validation, y_test = train_test_split(Zz, yy, test_size = 0.5, random_state = 0)
```

```python
#Training set Plot

for l, c, m,labl in zip(range(0,2), ('red', 'blue'), ('o','x'),('-ve','+ve')):
    plt.scatter(X_train[y_train == l, 0],X_train[y_train == l, 1],color=c,marker=m,label=labl)
    plt.legend()
    
    
plt.title('SVM Datsetset (training set)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

#Validation set Plot

for l, c, m,labl in zip(range(0,2), ('red', 'blue'), ('o','x'),('-ve','+ve')):
    plt.scatter(X_validation[y_validation == l, 0],X_validation[y_validation == l, 1],color=c,marker=m,label=labl)
    plt.legend()
    
    
plt.title('SVM Datsetset (Validation set)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()


#Testing set Plot

for l, c, m,labl in zip(range(0,2), ('red', 'blue'), ('o','x'),('-ve','+ve')):
    plt.scatter(X_test[y_test == l, 0],X_test[y_test == l, 1],color=c,marker=m,label=labl)
    plt.legend()
    
    
plt.title('SVM Datsetset (Testing set)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()



```

```python
#Create 2 SVM classifiers
#linear kernel

classifier = SVC(kernel='linear',random_state=0)
classifier.fit(X_train, y_train)



```

```python
#Best C for linear kernel
counter=0
yy = classifier.predict(X_test)

for i in range (1,100,5):
    
    classifier =SVC(kernel='linear',C=i,random_state=0)
    classifier.fit(X_validation, y_validation)
    
    f1 = f1_score(y_validation, yy)*100
    print("The f1 score : ",f1,"%")
    
   
    counter+=1
    print(counter)
    print("C=",i)
    
```
